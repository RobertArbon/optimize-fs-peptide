```python
import pandas as pd
import numpy as np
from osprey.config import Config
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import GPy

```

```python
df = Config('random/config-random.yaml').trial_results()

```

```python
df['m'] = df['parameters'].apply(lambda x: x['cluster__n_bins_per_feature'])
df = df.loc[(df['project_name']=='random_k2_log') & (df['m']>=20), :]
```

```python
y = df['mean_test_score'].values
# y_med = df['test_scores'].apply(lambda x: np.median(x)).values
X = df['parameters'].apply(lambda x: x['cluster__n_bins_per_feature']).values
```

```python
plt.scatter(np.log(X), y)
```

```python
X = X[:, np.newaxis]
y = y[:, np.newaxis]
```

```python
kernel = GPy.kern.Matern52(input_dim=1)
# kernel = GPy.kern.RBF(input_dim=1)
m = GPy.models.GPRegression(np.log(X[:,:]),y[:,:],kernel)
m.optimize_restarts(num_restarts = 20, verbose=False)
```

```python
m.plot()
plt.ylim((1.92,1.94))
```

```python
from scipy.stats import norm
def _ei(model, x):
    y_mean, y_var = model.predict(x)
    y_std = np.sqrt(y_var)
    y_best = model.Y.max(axis=0)
    z = (y_mean - y_best)/y_std
    result = y_std*(z*norm.cdf(z) + norm.pdf(z))
    return result
```

```python
np.random.seed(42)

y = df['mean_test_score'].values
X = df['parameters'].apply(lambda x: x['cluster__n_bins_per_feature']).values
idx = np.random.choice(np.arange(len(y)),20)
y = y[idx, np.newaxis]
X = X[idx, np.newaxis]

kernel = GPy.kern.Matern52(input_dim=1)
m = GPy.models.GPRegression(np.log(X),y,kernel)
_ = m.optimize_restarts(num_restarts = 20, verbose=False)
```

```python
x = np.arange(20,1001,10)[:,np.newaxis]
mu, var = m.predict(np.log(x))
ei = _ei(m, np.log(x))


lb = (mu - 1.96*np.sqrt(var)).reshape(-1)
ub = (mu + 1.96*np.sqrt(var)).reshape(-1)
x = x.reshape(-1)  
```

```python
from matplotlib.ticker import StrMethodFormatter

with sns.plotting_context('talk', font_scale=1):
    sns.set_style('whitegrid')
    
    fig, axes = plt.subplots(2, sharex=True)

    ax = axes[0]
    ax.plot(x, mu, label='Mean')
    ax.scatter(X,y, label='Data', c='k', marker='x')
    ax.fill_between(x,lb, ub, alpha=0.4, label='95% CI')
    ax.legend()
    ax.set_ylabel('VAMP-E')

    ax = axes[1]
    ax.plot(x,ei, label='EI')
    ax.scatter(x[np.argmax(ei)], np.max(ei), label='Suggestion')
    ax.set_xlabel('m')    
    ax.yaxis.set_major_formatter(StrMethodFormatter('{x:.1e}'))
    ax.set_ylabel('Expected Improvement')
    ax.legend()
    plt.tight_layout()
    plt.savefig('../figures/expected_improvement.png', transparent=True)

```
