```python
import seaborn as sns
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fracs = np.logspace(np.log10(0.005), np.log10(1), 5)
fracs
```

```python
paths = glob('data/100.0pc/*npy')
X = [np.load(x ) for x in paths]
```

```python
sample = np.vstack([x[:int(frac*len(x))] for x in X])
sample.shape
```

```python
# cols = sns.color_palette('light', len(fracs))
with sns.plotting_context('paper', font_scale=1.5):
    for i, frac in enumerate(fracs):
        sample = np.vstack([x[:int(frac*len(x))] for x in X])
        sns.distplot(sample, hist=False, label='{0:4.2f}$t_2$'.format(10*frac))
        plt.legend()
```
