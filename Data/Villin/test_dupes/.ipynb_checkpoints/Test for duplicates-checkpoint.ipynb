```python
traj = md.load('clone012-prot.xtc', top='top_prot.pdb')
```

```python
rmsd = np.ones(traj.n_frames-1)
import copy
ref = copy.copy(traj)
for idx in range(traj.n_frames-1):
    target = traj[idx+1]
    rmsd[idx] = md.rmsd(target, ref, frame=idx)
    
```
