```python
from osprey.trials import make_session, Trial
import pandas as pd
import numpy as np
```

```python
# Imports
from osprey.config import Config

# Load Configuation File
my_config = 'config.yaml'
config = Config(my_config)

# Retrieve Trial Results
df = config.trial_results()
db1 = config.trials()
history = db1.query(Trial).all() 



```

```python
df2 = pd.read_pickle('cse_trials_0-159.pickl')
print(df2.shape)
df3 = pd.read_pickle('cse_trials_160-end.pickl')
print(df3.shape)

df2 = pd.concat([df2, df3])
print(df2.shape)
df2.sort_values(by='id', inplace=True)
df2.head()
```

```python
def mean(x):
    try:
        return np.mean(x)
    except TypeError:
        return np.nan
    
```

```python
df2['mean_cse_test_score'] = df2['cse_test_scores'].apply(mean)
df2['mean_cse_train_score'] = df2['cse_train_scores'].apply(mean)

df2.head()
```

```python
df2['cse_status'] = ~(df2['mean_cse_test_score'].isnull() | df2['mean_cse_train_score'].isnull())
```

```python
print(df.columns)
print(df2.columns)
```

```python
df2.head()
```

```python
# Create new trial history.  
# Put the CSE results in place of the GMRQ results. 

# to_change = [('mean_test_score', 'cse_mean_test_score'), ('test_scores', 'cse_test_scores'), 
#             ('mean_train_score', 'cse_mean_train_score'), ('train_scores', 'cse_train_score')]
new_history = []
for trial in history: 
    # Check if it succeeded in both the CSE and the GMRQ-2 case. 
    if trial.status == 'SUCCEEDED':
        idx = df2['id']==trial.id
        if df2.loc[idx, 'cse_status'].values[0]:
            # Change the values
            trial.mean_test_score = df2.loc[idx, 'mean_cse_test_score'].values[0]
            trial.test_scores = df2.loc[idx, 'cse_test_scores'].values[0]
            
            trial.mean_train_score = df2.loc[idx, 'mean_cse_train_score'].values[0]
            trial.train_scores = df2.loc[idx, 'cse_train_scores'].values[0]
            
            new_history.append(trial)
        
```

```python
# Save new history to new database
project_names = df2['project_name'].unique()
for name in project_names:
    db2 = make_session('sqlite:///osprey-trials-cse.db', project_name=name)
    save_trials  = [t for t in new_history if t.project_name==name]
    
    # merge them into new db
    for trial in save_trials:
        db2.merge(trial)
    db2.commit()
    db2.close()


```

```python
# Load new database up and check results. 
my_config = 'config-cse.yaml'
config2 = Config(my_config)
df_new = config2.trial_results()
df_new.head()
```
