## Calculate mean GMRQ-2 and GMRQ-CSE


```python
def safe_mean(x):
    try:
        return np.array(x).mean()
    except TypeError:
        return np.nan

def safe_error(x):
    try:
        return np.array(x).std()
    except TypeError:
        return np.nan
```

```python
# Note - this assumes that the folds are all the same size.  This is fine for DHFR as all trajectories are 1000 ns long. 

observables = [ 'test_scores','cse_test_scores', 'cse_train_gaps']

for observable in observables:
#         df[observable+'_error'] = df.loc[:, observable].apply(safe_error)
        df[observable+'_mean'] = df.loc[:, observable].apply(safe_mean)
```

## FacetGrid with scores and gaps

```python
df4 = trials.loc[:, ['id','rank', 'cse_rank', 'cse_train_gaps', 'cse_train_n_timescales', 'cse_test_scores', 'test_scores']] 

```

```python
def f(x):
    low = np.floor(np.log10(x)).astype(int)
    high = low+1
    return r'${0} < \log(R) < {1}$'.format(low, high)
    
```

```python
df4['gap_cat'] = df4.groupby('id')['cse_train_gaps'].transform(np.mean).apply(f)

# df4['timescale_cat'] = df4['cse_train_gaps'].apply(lambda x: np.floor(np.log10(x)).astype(int).astype(str))
```

```python
df4.head()
```

```python
## Get the min & max of n_timescales  
df5 = df4.groupby('id')['cse_train_n_timescales'].aggregate([np.min, np.max])

df5['n_ts_width'] = df5['amax'] - df5['amin']
df5.reset_index(inplace=True)
df5.sort_values(by='id',inplace=True)
df5.rename(columns={'amin': 'n_ts_min', 'amax': 'n_ts_max'}, inplace=True)
df5 = df4.loc[:, ['id', 'rank', 'cse_rank']].drop_duplicates().merge(right = df5, on='id', how='inner')
df5.head()
```

```python
df4.sort_values(by='cse_rank', inplace=True, ascending=True)

df4.shape
```

```python
df4
```

```python
n_trials = 50
n_rows = 5*n_trials # 5 CV folds per trial
from matplotlib import ticker
cols = sns.color_palette('colorblind', 3)


with sns.plotting_context('paper', font_scale=2):
    sns.set_style('white')
    fig, axes = plt.subplots(4,2, sharex=True, figsize=(20,10))
    
    errbars = 0.8
    scale = 0.7
    for jdx, use_rank in enumerate(['rank', 'cse_rank']):
        df5.sort_values(by=use_rank, inplace=True, ascending=True)
        df4.sort_values(by=use_rank, inplace=True, ascending=True)
        ts_index = df5[use_rank] < n_trials
        row_index = df4[use_rank] < n_trials

        idx = 0  
#         bottom = df5.loc[ts_index, 'n_ts_min'].values
#         height = df5.loc[ts_index, 'n_ts_width'].values
#         left = df5.loc[ts_index, use_rank].values
#         axes[idx, jdx].bar(bottom=bottom, height=height, left=left, width=1, alpha=0.2)
#         axes[idx, jdx].scatter(df4.loc[row_index, use_rank]-0.25, df4.loc[row_index, 'cse_train_n_timescales'], s=10)
        sns.boxplot(x=use_rank, y='cse_train_n_timescales', ax=axes[idx, jdx], data=df4.loc[row_index, :], 
                    color=cols[-1])
        axes[idx, jdx].set_ylabel('# processes')
        axes[idx, jdx].set_xlabel("")
        axes[idx, jdx].set_title('Number of CSPs retained in training folds')


        idx += 1 
        sns.pointplot(x=use_rank, y='cse_train_gaps', hue='gap_cat', ax=axes[idx, jdx], data=df4.loc[row_index, :], join=False, 
                      legend=False, scale=scale, errwidth=errbars)
        axes[idx, jdx].legend(ncol=2)
        axes[idx, jdx].set_ylabel('Timescale ratio, R')
        axes[idx, jdx].set_xlabel("")
        axes[idx, jdx].set_title('Average timescale ratio, R, across training folds')
        axes[idx, jdx].set_yscale('log')


        idx += 1
        sns.pointplot(x=use_rank, y='cse_test_scores', hue='gap_cat', ax=axes[idx, jdx], data=df4.loc[row_index, :], join=False, 
                      legend=False, scale=scale, errwidth=errbars)
        axes[idx, jdx].legend(ncol=2)
        axes[idx, jdx].set_ylabel(r'GMRQ(CSP)')
        axes[idx, jdx].set_title('Mean GMRQ(CSP) across test folds')
        axes[idx, jdx].set_xlabel("")
        axes[idx, jdx].set_ylim((1.5,12.5))
        
        
        idx += 1
        sns.pointplot(x=use_rank, y='test_scores', hue='gap_cat', ax=axes[idx, jdx], data=df4.loc[row_index, :], join=False, 
                      scale=scale, errwidth=errbars)
        axes[idx, jdx].set_ylabel(r'GMRQ(2)')
        axes[idx, jdx].legend(ncol=2)
        axes[idx, jdx].set_title('Mean GMRQ(2) across test folds')
        axes[idx, jdx].set_ylim((2.2,3.2))


        axes[idx, jdx].xaxis.set_major_locator(ticker.MultipleLocator(10))
        axes[idx, jdx].xaxis.set_major_formatter(ticker.ScalarFormatter())


plt.tight_layout()
plt.savefig('trial_results.pdf')
```

## Correlation with optimized variables

```python
df6 = trials.groupby(['id']).agg({'test_scores':np.mean, 'cse_test_scores':np.mean})

df6.reset_index(inplace=True)
df6 = trials.loc[:, ['id', 'tica__lag_time','tica__n_components', 
                     'cluster__n_clusters', 'feature']].drop_duplicates().merge(df6, on='id', how='inner')
```

```python
df6['feature'].unique()
```

```python
rename = {'alpha_angle': r'$\alpha$', 
          'kappa_angle': r'$\kappa$',
          'psi-o_tor': r'$\psi-\omega$', 
         'pp_tor': r'$\psi-\phi$', 
         'psi_tor': r'$\psi$', 
         'phi-o_tor': r'$\phi-\omega$',
         'phi_tor': r'$\phi$', 
         'omega_tor': r'$\omega$'}
df6['feature'] = df6['feature'].apply(lambda x: rename[x])
df6.head()
```

```python
ylabels=['GMRQ(2)', 'GMRQ(CSP)']
variables = [ 'tica__n_components','tica__lag_time','cluster__n_clusters' ]
xlabels=['tICA components', 'tICA lag time', 'cluster centers' ]
with sns.plotting_context('paper', font_scale=2):
    sns.set_style('white')
    cols = sns.color_palette('colorblind',3)
    
    fig, axes = plt.subplots(4,2, figsize=(20,10))
    
    for jdx, score in enumerate(['test_scores', 'cse_test_scores']):
        idx=0
        sns.boxplot(x='feature', y=score, ax=axes[idx, jdx], data=df6)
        axes[idx, jdx].set_ylabel(ylabels[jdx])
        
        idx=1
        sns.boxplot(x=variables[idx-1], y=score, ax=axes[idx, jdx], data=df6, color=cols[0])
        axes[idx, jdx].set_ylabel(ylabels[jdx])
        axes[idx, jdx].set_xlabel(xlabels[idx-1])

        for idx in range(2,4):
            axes[idx, jdx].scatter(x=df6[variables[idx-1]], y=df6[score], alpha=0.1)
            axes[idx, jdx].set_ylabel(ylabels[jdx])
            axes[idx, jdx].set_xlabel(xlabels[idx-1])
 

        
plt.tight_layout()
plt.savefig('variable_vs_score.pdf')
```
