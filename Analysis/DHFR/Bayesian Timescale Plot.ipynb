```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
df = pd.read_pickle('bayesian_timescales.pickl')
```

```python
df.shape
```

```python
fig, ax = plt.subplots(1, figsize=(10,10))

ax = sns.pointplot(x='id',hue='Process',y='Timescale', estimator=np.median, data=df.loc[df['Process']<11,:], join=False, dodge=True, 
                   ci=100, scale=.5, )
ax.set_yscale('log')
```

```python
df.groupby(['Process', 'id']).aggregate([np.mean,np.median, np.std, np.min, np.max])
```

```python
g = sns.FacetGrid(data=df, row='Process', hue='id', sharey=False
                 )
g.map(sns.distplot, 'Timescale', hist=False)
for ax in g.axes.flatten():
    ax.set_xscale('log')
```
