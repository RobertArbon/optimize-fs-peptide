```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
from collections import OrderedDict
```

```python
with sns.plotting_context('paper', font_scale=1.25):
    np.random.seed(42)
    x = np.linspace(-2, 10, 20)
    y = -0.1*x*x + x 
    y_err = y+ np.random.normal(0, 1, len(y))
    plt.plot(x, y)
    plt.scatter(x, y_err)
    plt.xlabel('x')
    plt.ylabel('y')
    
    max_y_er = np.max(y_err)
    max_x_er = x[np.argmax(y_err)]
    
    xlim = plt.gca().get_xlim()
    ylim = plt.gca().get_ylim()
    
    plt.hlines(max_y_er, xlim[0], max_x_er)
    plt.vlines(max_x_er, ylim[0], max_y_er)
    
    max_y = np.max(y)
    max_x = x[np.argmax(y)]
    
    plt.hlines(max_y, xlim[0], max_x)
    plt.vlines(max_x, ylim[0], max_y)
    
    plt.xlim(xlim)
    plt.ylim(ylim)
    
    
    
```
